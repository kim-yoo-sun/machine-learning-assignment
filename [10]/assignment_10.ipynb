## define the feature functions

- feature vector is defined by $(1, f_1(x, y), f_2(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_feature(point):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    feature = np.array([np.ones(point[:,0].shape), point[:,0], point[:,1], point[:,0] * point[:,0], point[:,1] * point[:,1], point[:,0] * point[:,1], point[:,0] * point[:,0] * point[:,0], point[:,1] * point[:,1] * point[:,1], point[:,1] * point[:,1] * point[:,1] * point[:,1] * point[:,1], point[:,0] * point[:,1] * point[:,1] * point[:,1] * point[:,1], point[:,0] * point[:,0] * point[:,1] * point[:,1] * point[:,1], point[:,0] * point[:,0] * point[:,0] * point[:,1] * point[:,1]])
    feature = feature.T
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return feature
```

## define the linear regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature = $(1, f_1(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_linear_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = np.matmul(theta, feature.T)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    
    return value
```

## define sigmoid function with input

- $z \in \mathbb{R}$

```python
def sigmoid(z):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = 1 / (1 + np.exp(-z))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value 
```

## define the logistic regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$

```python
def compute_logistic_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = sigmoid(compute_linear_regression(theta, feature))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value
```

## define the residual function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_residual(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    label = label.squeeze()
    residual = residual = (- 1) * (label * np.log1p(compute_logistic_regression(theta, feature)) + (np.ones(label.shape[0])-label) * (np.log1p(np.ones(label.shape[0])-compute_logistic_regression(theta, feature))))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return residual
```

## define the loss function for the logistic regression

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_loss(theta, feature, label, alpha):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    label = label.squeeze()
    loss = np.sum(compute_residual(theta, feature, label)) / (label.shape[0]) + (alpha / 2) * np.inner(theta, theta)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return loss
```

## define the gradient of the loss with respect to the model parameter $\theta$

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_gradient(theta, feature, label, alpha):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    label = label.squeeze()
    gradient = (np.matmul(np.multiply(np.ones(label.shape[0]) - label, compute_logistic_regression(theta, feature)), feature) + np.matmul(np.multiply(label, compute_logistic_regression(theta, feature) - np.ones(label.shape[0])), feature)) / label.shape[0] + alpha * theta
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return gradient
```

## compute the accuracy of the prediction for point with a given model parameter

```python
def compute_accuracy(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    correct_count = 0
    total_count = 0
    value = compute_logistic_regression(theta, feature)
    
    for i in range(label.shape[0]):
        if value[i] >= 0.5:
            total_count = total_count + 1
            if label[i] == 1:
                correct_count = correct_count + 1
        else:
            total_count = total_count + 1
            if label[i] == 0:
                correct_count = correct_count + 1
    
    accuracy = correct_count / total_count
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return accuracy
```

## initialize the gradient descent algorithm

```python
number_iteration    = 22430 # you can change this value as you want
learning_rate       = 0.3 # you can change this value as you want
number_feature      = 12 # you can change this value as you want
alpha               = 0.00003 # you can change this value as you want

theta                       = np.zeros(number_feature)
loss_iteration_train        = np.zeros(number_iteration)
loss_iteration_test         = np.zeros(number_iteration)
accuracy_iteration_train    = np.zeros(number_iteration)
accuracy_iteration_test     = np.zeros(number_iteration)
```

## run the gradient descent algorithm to optimize the loss function with respect to the model parameter

```python
for i in range(number_iteration):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
 
    theta           = theta - learning_rate * compute_gradient(theta, compute_feature(data_train), data_train[:,2], alpha)
    loss_train      = compute_loss(theta, compute_feature(data_train), data_train[:,2], alpha)
    loss_test       = compute_loss(theta, compute_feature(data_test), data_test[:,2], alpha)
    accuracy_train  = compute_accuracy(theta, compute_feature(data_train), data_train[:,2])
    accuracy_test   = compute_accuracy(theta, compute_feature(data_test), data_test[:,2])

    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    loss_iteration_train[i]     = loss_train
    loss_iteration_test[i]      = loss_test
    accuracy_iteration_train[i] = accuracy_train
    accuracy_iteration_test[i]  = accuracy_test

theta_optimal = theta
```
