## define the feature functions

- feature vector is defined by $(1, f_1(x, y), f_2(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_feature1(point):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    feature = np.array([np.ones(point[:,0].shape), point[:,0], point[:,1], point[:,0] * point[:,0], point[:,1] * point[:,1], point[:,0] * point[:,1], point[:,0] * point[:,0] * point[:,0], point[:,0] * point[:,0] * point[:,1], point[:,0] * point[:,1] * point[:,1], point[:,1] * point[:,1] * point[:,1], np.cos(point[:,0]), np.cos(point[:,1])])
    feature = feature.T
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return feature
```

```python
def compute_feature2(point):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    feature = np.array([np.ones(point[:,0].shape), point[:,0], point[:,1], point[:,0] * point[:,0], point[:,1] * point[:,1], point[:,0] * point[:,1], point[:,0] * point[:,0] * point[:,0], point[:,0] * point[:,0] * point[:,1], point[:,0] * point[:,1] * point[:,1], point[:,1] * point[:,1] * point[:,1]])
    feature = feature.T
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return feature
```

## define the linear regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature = $(1, f_1(x, y), \cdots, f_{k-1}(x, y)) \in \mathbb{R}^k$

```python
def compute_linear_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = np.matmul(theta, feature.T)
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value
```

## define sigmoid function with input

- $z \in \mathbb{R}$

```python
def sigmoid(z):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = 1 / (1 + np.exp(-z))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value 
```

## define the logistic regression function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$

```python
def compute_logistic_regression(theta, feature):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    value = sigmoid(compute_linear_regression(theta, feature))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return value
```

## define the residual function

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_residual(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    label = label.squeeze()
    residual = residual = (- 1) * (label * np.log1p(compute_logistic_regression(theta, feature)) + (np.ones(label.shape[0])-label) * (np.log1p(np.ones(label.shape[0])-compute_logistic_regression(theta, feature))))
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return residual
```

## define the loss function for the logistic regression

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_loss(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    label = label.squeeze()
    loss = np.sum(compute_residual(theta, feature, label)) / (label.shape[0])
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return loss
```

## define the gradient of the loss with respect to the model parameter $\theta$

- $\theta = (\theta_0, \theta_1, \cdots, \theta_{k-1}) \in \mathbb{R}^k$
- feature $= (1, f_1(x, y), \cdots, f_{k-1}(x, y) \in \mathbb{R}^k$
- label $= l \in \{0, 1\}^k$

```python
def compute_gradient(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    label = label.squeeze()
    gradient = (np.matmul(np.multiply(np.ones(label.shape[0]) - label, compute_logistic_regression(theta, feature)), feature) + np.matmul(np.multiply(label, compute_logistic_regression(theta, feature) - np.ones(label.shape[0])), feature)) / label.shape[0]
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return gradient
```

## compute the accuracy of the prediction for point with a given model parameter

```python
def compute_accuracy(theta, feature, label):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    correct_count = 0
    total_count = 0
    value = compute_logistic_regression(theta, feature)
    
    for i in range(label.shape[0]):
        if value[i] >= 0.5:
            total_count = total_count + 1
            if label[i] == 1:
                correct_count = correct_count + 1
        else:
            total_count = total_count + 1
            if label[i] == 0:
                correct_count = correct_count + 1
    
    accuracy = correct_count / total_count
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    return accuracy
```

## initialize the gradient descent algorithm

```python
data1_number_iteration   = 635500
data2_number_iteration   = 335500

data1_learning_rate   = 0.3 
data2_learning_rate   = 0.3 

data1_number_feature    = 12
data2_number_feature    = 10

theta1  = np.zeros(data1_number_feature)
theta2  = np.zeros(data2_number_feature)

data1_loss_iteration  = np.zeros(data1_number_iteration)
data2_loss_iteration  = np.zeros(data2_number_iteration)

data1_accuracy_iteration    = np.zeros(data1_number_iteration)
data2_accuracy_iteration    = np.zeros(data2_number_iteration)
```

## run the gradient descent algorithm to optimize the loss function with respect to the model parameter

```python
for i in range(data1_number_iteration):

    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    theta1      = theta1 - data1_learning_rate * compute_gradient(theta1, compute_feature1(data1), data1[:,2])
    loss1       = compute_loss(theta1, compute_feature1(data1), data1[:,2])
    accuracy1   = compute_accuracy(theta1, compute_feature1(data1), data1[:,2])
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    data1_loss_iteration[i]     = loss1
    data1_accuracy_iteration[i] = accuracy1

data1_theta_optimal = theta1
```

```python
for i in range(data2_number_iteration):
    
    # ++++++++++++++++++++++++++++++++++++++++++++++++++
    # complete the blanks
    #
    theta2      = theta2 - data2_learning_rate * compute_gradient(theta2, compute_feature2(data2), data2[:,2])
    loss2       = compute_loss(theta2, compute_feature2(data2), data2[:,2])
    accuracy2   = compute_accuracy(theta2, compute_feature2(data2), data2[:,2])
    #
    # ++++++++++++++++++++++++++++++++++++++++++++++++++

    data2_loss_iteration[i]     = loss2
    data2_accuracy_iteration[i] = accuracy2

data2_theta_optimal = theta2
```
